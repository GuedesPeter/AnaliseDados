```python
altura = np.loadtxt('/content/drive/MyDrive/Aulas SENAC/AD2023/Scripts/dados/alturaCenso.csv')


print(np.std(altura))

k = 1 + (10/3)*np.log10(altura.size)
k = int(k)

print(k)

print('Valor Minimo: ', altura.min())
print('Valor Maximo: ', altura.max())
print('Amplitude: ',  altura.max()-altura.min())

rotulos = ['A', 'B', 'C', 'D', 'F', 'G', 'H', 'I', 'J', 'K', 'L']

frequencia = pd.value_counts(
    pd.cut(
        x=altura,
        bins=k,
        labels=rotulos
    )
)

print('Freq. rotulada: ', frequencia)
print('Freq. rotulada ordenada: ', frequencia.sort_index())
```
